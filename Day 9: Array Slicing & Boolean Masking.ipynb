```python
temperatures = np.round(np.random.uniform(20.00,100.00, size=50),2)

def sorted_temperatures():
    sliced_temperatures = temperatures[:10]

    mask = temperatures > 80
    Heatwaves = temperatures[mask]

    count_heatwaves=np.size(Heatwaves)

    Inbetween_temperatures = (temperatures >50) & (temperatures <60)
    new_temperatures = temperatures[Inbetween_temperatures]



print(f"First 10 temperature recodings are : {sliced_temperatures}")
print(f"Temperatures that can be counted as Heatwaves: {Heatwaves}")
print(f"Count of heatwaves: {count_heatwaves}")
print(f"Temperatures that can be counted as Inbetween: {new_temperatures}")

```
