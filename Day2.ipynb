```python
#list of names to be corrected

names = ["  alIcE", "BOB  ", "chaRLie", " dAvid "]
cleaned_emails = []

#The for loop to cleanup the name
for name in names:
  cleaned_name = name.strip().lower()
  email = cleaned_name + "@company.com"
  cleaned_emails.append(email)

print(cleaned_emails)
```
