```python
portfolio = {
    "AAPL": 270.23,
    "TSLA": 400.62,
    "MSFT": 422.79,
    "GOOGL": 341.68,
}

# Define the target price for stocks
targetprice =150

# Iterate through the stock prices in the portfolio
# Determine whether to BUY, SELL, or HOLD based on the target price

for stock,price in portfolio.items():

  # If the price is more than 10% below the target price, recommend BUY

  if price < targetprice * 0.9:
    print(f'{stock}: BUY')

  # If the price is more than 10% above the target price, recommend SELL

  elif price > targetprice * 1.1:
    print(f'{stock}: SELL')

  # Otherwise (within 10% of target price), recommend HOLD

  else:
    print("HOLD")

# Demonstrate safe access for a non-existent stock ('AMZN')
# Returns 0.0 if the key is not found, instead of raising a KeyError

print(f"AMZN price : {portfolio.get("AMZN",0.0)}")
```
