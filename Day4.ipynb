# Sales Data Cleaning and Analysis

```python
raw_sales = [
    {"product": "  iPhone ", "price": 999, "user": "buyer1"},
    {"product": "CasE", "price": 25, "user": "TEST"},
    {"product": " MacBooK  ", "price": 1500, "user": "buyer2"},
    {"product": "AirPods", "price": None, "user": "buyer3"}, # Missing price!
    {"product": " charger ", "price": 30, "user": "buyer4"}
]

sales =[]
total_revenue = 0 # Initialize total revenue

for sale in raw_sales:
  price = sale.get("price")
  if price is None:
    continue

  cleaned_product = sale["product"].strip().lower() # Clean product name to lowercase
  cleaned_user = sale["user"].strip().lower()

  if cleaned_user == "test": # Filter out sales from 'TEST' user
    continue

  sales.append({"product" : cleaned_product , "price" : sale["price"], "user" : cleaned_user})
  total_revenue += sale["price"] # Add to total revenue

print("Cleaned Sales:", sales)
print("Total Revenue:", total_revenue)
```
