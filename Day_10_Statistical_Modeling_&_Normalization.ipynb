```python
import numpy as np

# Generate: Create an array of 100 random "House Prices" between $100,000 and $1,000,000.
House_prices = np.random.randint(100000, 1000000, 100)

# Analyze: Print the Mean and Standard Deviation.
mean = np.mean(House_prices)
std_deviation = np.std(House_prices)
print(f"Mean of House Prices: {mean:.2f}")
print(f"Standard Deviation of House Prices: {std_deviation:.2f}")

# Filter: Use a Boolean Mask to find "Luxury Houses" (prices that are greater than the Mean + (1.5 * Std_Dev)).
luxury_threshold = mean + (1.5 * std_deviation)
luxury_mask = House_prices > luxury_threshold
luxury_houses = House_prices[luxury_mask]
print(f"\nNumber of Luxury Houses: {len(luxury_houses)}")
print(f"Luxury Houses (prices > {luxury_threshold:.2f}):")
print(luxury_houses)

# Normalize: Create a new array called scaled_prices where all house prices are scaled between 0 and 1.
min_price_all = np.min(House_prices)
max_price_all = np.max(House_prices)
scaled_prices = (House_prices - min_price_all) / (max_price_all - min_price_all)

print("\nScaled Prices (all houses):")
print(scaled_prices)

# Verify: Print the min and max of your scaled_prices to ensure they are exactly 0.0 and 1.0.
Min_scaled_price = np.min(scaled_prices)
Max_scaled_price = np.max(scaled_prices)

print(f"\nMinimum Scaled Price: {Min_scaled_price:.1f}")
print(f"Maximum Scaled Price: {Max_scaled_price:.1f}")
```
