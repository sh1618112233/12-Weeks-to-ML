```python
import pandas as pd
```

```python
# This cell creates a Pandas DataFrame, calculates 'Total Value' for each stock,
# and then filters and sorts the DataFrame based on 'Total Value'.
data = {
    "Ticker": ["AAPL", "TSLA", "GOOGL", "AMZN", "MSFT"],
    "Price": [175.50, 240.20, 150.10, 185.00, 420.30],
    "Shares": [10, 5, 20, 8, 15]
}

df = pd.DataFrame(data)
df["Total Value"] = df["Price"] * df["Shares"]
df.describe()
df_sorted = df.sort_values(by = "Total Value")
df_sorted[df_sorted["Total Value"] > 1000]

df
```
