```python
import pandas as pd
import numpy as np

# Load the dataset from a Google Sheet CSV export URL
# The URL is structured to directly export the sheet as a CSV file.
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1-Zj-SKsyq65HKkHmyvtgWqhRiIi1WvK_nELZJ76S-xU/export?format=csv&gid=1802847773')
# Display the initial DataFrame to inspect its structure and content
df

# Replace specific string values, like '[EMPTY]', with NumPy's NaN (Not a Number).
# This standardizes missing value representation for easier handling.
df.replace(['[EMPTY]'], np.nan, inplace=True)
# Check the count of missing values for each column after replacement.
df.isna().sum()

# Drop rows where the 'Customer Name' column has missing (NaN) values.
# 'inplace=True' modifies the DataFrame directly.
df.dropna(subset=["Customer Name"], inplace=True)

# Clean and convert the 'Amount' column:
# 1. Convert 'Amount' to string type to apply string methods.
# 2. Remove '$' and ',' characters using string replacement.
df['Amount'] = df['Amount'].astype(str).str.replace('$', '',regex=False).str.replace(',','', regex=False)
# 3. Convert the cleaned 'Amount' column to numeric, coercing errors to NaN.
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
# 4. Fill any remaining NaN values in the 'Amount' column with the string 'na'.
#    Note: This will change the column's dtype to object if it contains non-numeric values.
df['Amount'] = df['Amount'].fillna('na')
# Display the final processed DataFrame.
df
```
