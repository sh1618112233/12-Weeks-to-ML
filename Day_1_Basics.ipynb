```python
# Script to calculate monthly expenses and budget
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: "))
rent = float(input("Enter your rent: "))
Groceries = float(input("Enter your groceries: "))
balance = monthly_income - rent - Groceries
print(f"Hi {name}, after expenses, you have ${balance:.2f} left.")
```
