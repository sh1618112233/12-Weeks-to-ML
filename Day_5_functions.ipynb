```python
def process_sales(data, tax=0.05):
  """
  Processes a list of sales data, cleaning and filtering entries, and calculating total revenue.

  Args:
    data (list): A list of dictionaries, where each dictionary represents a sale
                 and should contain 'product', 'price', and 'user' keys.
    tax (float, optional): The tax rate to apply to the total revenue. Defaults to 0.05 (5%).

  Returns:
    tuple: A tuple containing:
      - list: A list of dictionaries representing the cleaned and filtered sales.
              Each dictionary includes 'product', 'price', and 'user' keys.
              Product and user names are cleaned (stripped and lowercased).
      - float: The final total revenue after applying the tax.
  """
  # Initialize an empty list to store cleaned sales data
  sales=[]
  # Initialize total revenue before processing sales
  total_revenue=0

  # Iterate through each sale entry in the provided data list
  for sale in data:
    # Get the price, using .get() to avoid KeyError if 'price' is missing
    price = sale.get("price")
    # If price is None, skip this sale entry
    if price is None:
      continue

    # Clean the product name: remove leading/trailing whitespace and convert to lowercase
    cleaned_product=sale["product"].strip().lower()
    # Clean the user name: remove leading/trailing whitespace and convert to lowercase
    cleaned_user=sale["user"].strip().lower()

    # Filter out sales made by a user named 'test'
    if cleaned_user == "test":
      continue

    # Add the cleaned and valid sale to the sales list
    sales.append({"product" : cleaned_product , "price" : sale["price"] , "user" : cleaned_user})

    # Add the price of the current sale to the total revenue
    total_revenue += price

  # Calculate the final revenue by applying the tax rate
  final_revenue = total_revenue * (1+tax)

  # Return the cleaned sales list and the final calculated revenue
  return sales , final_revenue
```

```python
data1 = [
    {"product": "  iPhone ", "price": 999, "user": "buyer1"},
    {"product": "CasE", "price": 25, "user": "TEST"},
    {"product": " MacBooK  ", "price": 1500, "user": "buyer2"},
    {"product": "AirPods", "price": None, "user": "buyer3"}, # Missing price!
    {"product": " charger ", "price": 30, "user": "buyer4"}
]
```

```python
sales, final_revenue = process_sales(data1)

print(f"The cleaned sales data is: {sales}")
print(f"The final revenue is: {final_revenue}")
```
