```python
import numpy as np

stock_prices = np.random.uniform(100, 500, size=10)
print("Original Stock Prices:", stock_prices)

percentage_change = np.random.uniform(-0.05, 0.10, size=10)
multiplier = 1 + percentage_change
print("Percentage Change (as multiplier):\n", multiplier)

new_prices = stock_prices * multiplier
print("New Stock Prices:", new_prices)

max_price = np.max(new_prices)
min_price = np.min(new_prices)
mean_price = np.mean(new_prices)

print(f"Maximum New Price: {max_price:.2f}")
print(f"Minimum New Price: {min_price:.2f}")
print(f"Mean New Price: {mean_price:.2f}")
```
