```python
def password_strength(password):
  """
  Calculates the strength score of a given password.

  The score is based on the following criteria:
  - 1 point if the password length is 8 or more characters.
  - 1 point if it contains at least one uppercase letter.
  - 1 point if it contains at least one lowercase letter.
  - 1 point if it contains at least one digit.
  - 1 point if it contains at least one special character (!@#$%^&*()).

  Args:
    password (str): The password string to evaluate.

  Returns:
    int: The strength score of the password (maximum 5).

  Example:
    >>> password_strength("Lyryt787878*")
    5
    >>> password_strength("yyryt")
    1
  """
  score = 0
  special_chars = "!@#$%^&*()"

  if len(password) >= 8:
    score += 1

  has_upper = False
  has_lower = False
  has_digit = False
  has_special = False

  for char in password:
    if char.isupper():
      has_upper = True
    elif char.islower():
      has_lower = True
    elif char.isdigit():
      has_digit = True
    elif char in special_chars:
      has_special = True

  if has_upper:
    score += 1
  if has_lower:
    score += 1
  if has_digit:
    score += 1
  if has_special:
    score += 1

  return score


print(password_strength(password))
```
