```python
grades_db = {}

def display_menu():
    print("\n--- Grade Management Menu ---")
    print("1. Add Student")
    print("2. View All Grades")
    print("3. Calculate Average")
    print("4. Exit")

while True:
    display_menu()
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        name = input("Enter student name: ")
        try:
            grade = float(input(f"Enter grade for {name}: "))
            grades_db[name] = grade
            print(f"Grade for {name} added successfully.")
        except ValueError:
            print("Invalid grade. Please enter a number.")
    elif choice == '2':
        if not grades_db:
            print("No students in the database yet.")
        else:
            print("\n--- All Grades ---")
            for name, grade in grades_db.items():
                print(f"Student: {name}, Grade: {grade}")
    elif choice == '3':
        if not grades_db:
            print("No students in the database to calculate average.")
        else:
            total_grades = sum(grades_db.values())
            average_grade = total_grades / len(grades_db)
            print(f"\nAverage grade: {average_grade:.2f}")
    elif choice == '4':
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```

```python
grades_db = {}

def menu():
  print("\n------Student grade tracer------\n")
  print("1. Add student")
  print("2. View all grades")
  print("3. Calculate average")
  print("4. Top performer")
  print("5. Exit")

while True:
  menu() # Call the menu function to display options

  choice = input("Enter your choice (1-5): ")

  if choice.isdigit():
    choice_int = int(choice)

    if choice_int == 1:
      name = input("Enter student name: ")
      try:
        grade = float(input(f"Enter grade for {name}: "))
        grades_db[name] = grade
        print(f"Grade for {name} entered successfully")

      except ValueError:
        print(f"Incorrect grade entered for {name}")


    elif choice_int == 2:
      if not grades_db:
        print("No students found")
      else:
        for name, grade in grades_db.items():
          print(f"Student: {name}, Grade: {grade}")

    elif choice_int == 3:
      if not grades_db:
        print("No students found")
      else:
        total_grades = sum(grades_db.values())
        average_grade = total_grades / len(grades_db)
        print(f"Average grade: {average_grade}")

    elif choice_int == 4:
      if not grades_db:
        print("No students found to determine top performer.")
      else:
        top_performer =max(grades_db, key=grades_db.get)
        print(f"Top performer: {top_performer}")


    elif choice_int == 5:
      print("Exiting program. Goodbye")
      break

    else:
      print("Invalid choice. Please enter a number between 1 and 5")
  else:
    print("Invalid choice. Please enter a number between 1 and 5")
```
